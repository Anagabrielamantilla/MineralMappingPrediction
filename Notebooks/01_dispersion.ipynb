```python
#Importar los datos 

df=pd.read_csv('/content/MineralPrediction/Data/00_Training.csv') 
```

```python
# Visualizar la tabla de datos 

df.head(2)
```

<image src="https://drive.google.com/uc?export=view&id=14t_xuo49e4cjV4Dtzlkx6kMYSKfPlDZy" width=5%>  **Graficar los datos**

```python
# Generar el gráfico de dispersión entre la variable X1 y X2

plt.scatter(df['X1'],df['X2'], c=df['DEP'])
plt.colorbar()
```

```python
# Generar el gráfico de dispersión 3D entre las variables X8,X9 y X10
# Para calcular el de otras variables solo deben cambiarse los nombres en color rojo, que corresponden al nombre de las columnas de la tabla

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(df['X8'],df['X9'], df['X10'], c=df['DEP'])
```
