```python
# Importar las librerías 

import seaborn as sn  #usada para graficar la matriz de correlación
from sklearn.preprocessing import StandardScaler #usada para estandarizar los datos
import pandas as pd #usada para cargar los datos delimitados por comas (.csv)
```

```python
#Importar los datos 

df = pd.read_csv('/content/MineralPrediction/Data/00_Training.csv')
```

```python
# Visualizar la tabla de datos 

df.head()
```

<image src="https://drive.google.com/uc?export=view&id=14t_xuo49e4cjV4Dtzlkx6kMYSKfPlDZy" width=5%>  **Calcular coeficiente de correlación de Pearson**

```python
# Definir los nombres de las columnas que se van a usar

columns = ['X1','X2','X3','X4','X5','X6','X7','X8','X9','X10','X11']
X = df[columns]
```

```python
# Estandarizar las variables 

XX=df[columns].values
x = StandardScaler().fit_transform(XX)
```

```python
# Llevar x (np) a un df
# Este procedimiento es para cambiar el tipo de dato. De esta forma se podrá graficar la matriz de correlación

X2=pd.DataFrame(data=x,columns=columns)
```

```python
# Graficar la matriz de correlación

corrMatrix = X.corr()
sn.set(rc = {'figure.figsize':(14,11)}, font_scale=1.1)
sn.heatmap(corrMatrix, annot=True,cmap="jet", alpha= 1)
```
